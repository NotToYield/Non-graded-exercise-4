```python
import random
import time

# Define a Card class to represent each card
class Card:
    def __init__(self, name, health, attack, defense):
        self.name = name
        self.health = health
        self.attack = attack
        self.defense = defense

    def __str__(self):
        return f"Name: {self.name}, Health: {self.health}, Attack: {self.attack}, Defense: {self.defense}"

# Function to simulate the attack process
def attack(player_card, opponent_card):
    print(f"\n{player_card.name} attacks {opponent_card.name}!")

    # If player's attack is greater than the opponent's health, the opponent's card is defeated
    if player_card.attack >= opponent_card.health:
        print(f"{opponent_card.name} has been defeated!")
        return True
    
    # Otherwise, apply damage considering the opponent's defense
    if player_card.attack >= opponent_card.defense:
        damage_to_health = player_card.attack - opponent_card.defense
        opponent_card.defense = 0
        opponent_card.health -= damage_to_health
        print(f"{opponent_card.name}'s defense has been broken! It lost {damage_to_health} health.")
    else:
        opponent_card.defense -= player_card.attack
        print(f"{opponent_card.name}'s defense reduced to {opponent_card.defense}.")

    return opponent_card.health <= 0

# Initialize the card deck
card_deck = [
    Card("Diablo", 100, 90, 60),
    Card("Medusa", 100, 70, 70),
    Card("Jester", 120, 60, 90),
    Card("Troll", 150, 40, 94),
    Card("Specter", 100, 70, 70),
    Card("Mist", 100, 75, 65),
    Card("Savage", 100, 90, 50),
    Card("Marauder", 100, 85, 50),
    Card("Wimp", 110, 40, 85),
    Card("Sorcerer", 100, 70, 55),
]

# Number of cards each player will have in their deck
deck_size = 3

# Draw random cards for player and opponent
player_deck = random.sample(card_deck, deck_size)
opponent_deck = random.sample(card_deck, deck_size)

# Game loop
while player_deck and opponent_deck:
    print("\nYour deck:")
    for idx, card in enumerate(player_deck):
        print(f"{idx + 1}: {card}")
    
    opponent_card = random.choice(opponent_deck)
    print(f"\nOpponent's card: {opponent_card}")

    choice = int(input("\nPick a card to fight (1-3): ")) - 1
    player_card = player_deck[choice]

    # Player's turn
    print(f"\nYou chose {player_card.name}!")
    if attack(player_card, opponent_card):
        print(f"{opponent_card.name} has been removed from the opponent's deck.")
        opponent_deck.remove(opponent_card)
    time.sleep(3)

    if not opponent_deck:
        print("\nCongratulations! You've defeated all opponent's cards. You win!")
        break

    # Opponent's turn
    print("\nOpponent is choosing a card...")
    time.sleep(2)
    
    opponent_card = random.choice(opponent_deck)
    print(f"\nOpponent chose {opponent_card.name}!")
    
    print("\nYour deck:")
    for idx, card in enumerate(player_deck):
        print(f"{idx + 1}: {card}")
    
    choice = int(input("\nPick a card to defend (1-3): ")) - 1
    player_card = player_deck[choice]

    print(f"\nOpponent's {opponent_card.name} attacks your {player_card.name}!")
    if attack(opponent_card, player_card):
        print(f"{player_card.name} has been removed from your deck.")
        player_deck.remove(player_card)
    time.sleep(3)

    if not player_deck:
        print("\nAll your cards have been defeated. You lose!")
        break

print("\nGame Over!")
```
